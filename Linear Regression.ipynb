```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  Cargar datos
from sklearn.datasets import load_boston
boston = load_boston()
```

Formula de la función de coste para minimizar el error. Error cuadrático medio vectorial: $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
# Crear matrices inicailes
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)
```

```python
# Representar nube de puntos
plt.scatter(X, Y, alpha=0.3)

# Añadir columna de 1s, para término independiente
X = np.array([np.ones(506), X]).T

# Calcular función beta (función de coste)
B = np.linalg.inv(X.T @ X) @ X.T @ Y

# Representar línea de mínimo error
plt.plot([4, 9], [ B[0] + B[1]*4 , B[0] + B[1]*9 ], c="red")
plt.show()
```
